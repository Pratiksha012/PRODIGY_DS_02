```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('/Users/pratiksha/Documents/Internships/Prodigy Infotech/Task 3 Dataset - Titanic/train.csv')
print(df)
```

```python
print(df.info())
```

```python
print("Missing Values:")
print(df.isnull().sum())
df["Age"].fillna(df["Age"].mean(), inplace=True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace=True)
```

```python
# EDA - Exploring relationships between variables
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Pclass")
plt.show()
```

```python
# Visualizing survival by Sex
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Sex")
plt.show()
```

```python
# Visualizing age distribution
sns.histplot(df["Age"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
# Visualizing fare distribution
sns.histplot(df["Fare"], bins=30, kde=True)
plt.title("Fare Distribution")
plt.show()
```

```python
# Visualizing survival by Embarked
sns.countplot(x="Embarked", hue="Survived", data=df)
plt.title("Survival by Embarked")
plt.show()
```

```python
# Correlation matrix
numeric_df = df.select_dtypes(include=['number'])
correlation_matrix = numeric_df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
